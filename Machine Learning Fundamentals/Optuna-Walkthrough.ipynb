```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
from pandas import read_csv

dataset = read_csv("/workspaces/Main-DSDT-Example-Repo/Machine Learning Fundamentals/no_null_encoded_titanic (1) (1).csv")
```

```python
display(dataset)
```

```python
target_column_name = "survived"
```

```python
inputs = dataset.drop(columns = [target_column_name, 'alive'])
```

```python
labels = dataset[target_column_name]
```

```python

training_inputs, test_inputs, training_labels, test_labels = train_test_split(inputs, labels)
```

```python
# defining what a decision tree classifier is for the computer
from sklearn.tree import DecisionTreeClassifier
```

```python
# define the objective function for the computer
# this tells the computer what to do for each experiment
def titanic_trial_function(param_suggestor):
    
    # define the list of possible categories for criterion
    category_list = ['gini', 'entropy', 'log_loss']

    # generate the suggestions
    criterion_suggestion = param_suggestor.suggest_categorical('criterion', category_list) # categorical suggestion
    max_depth_suggestion = param_suggestor.suggest_int('max depth', low = 1, high = 10) # integer suggestion
    min_leaf_suggestion = param_suggestor.suggest_float('min leaf',  low = 0, high = 1) # float suggestion


    # create the actual trial model
    trial_model = DecisionTreeClassifier(
        criterion=criterion_suggestion, 
        max_depth=max_depth_suggestion, 
        min_samples_leaf=min_leaf_suggestion
        )

    # train the model before we can make predictions
    trial_model.fit(training_inputs, training_labels)

    # get the actual predictions from the model so we can get the score
    trial_predictions = trial_model.predict(test_inputs)

    # get the final score of the trial model
    score = accuracy_score(test_labels, trial_predictions)

    # send back the score of the model in this trial
    return score
```

```python
# send the computer to the Optuna library to understand the functions
import optuna

# create the study - BUT DOES NOT ACTUALLY EXECUTE THE STUDY
study_titanic = optuna.create_study(direction = 'maximize')

# Actually tell the computer to execute the study - do some actual experiments
# pass in our objective function
study_titanic.optimize(titanic_trial_function, timeout = 6)

# get the best hyperparameters
best_hyper_parameters = study_titanic.best_params

# display the best hyperparameters
display("best params for titanic", best_hyper_parameters)
```
