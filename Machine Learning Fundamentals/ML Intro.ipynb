```python
from pandas import read_csv

dataset = read_csv("/workspaces/Main-DSDT-Example-Repo/Machine Learning Fundamentals/no_null_encoded_titanic (1) (1).csv")
```

```python
dataset = dataset.drop_duplicates()
```

```python
display(dataset)
```

```python
target_column_name = 'survived'


target_column = dataset[target_column_name]

display(target_column)
```

```python
input_feature_set = dataset.drop(columns = ['survived', 'alive'])

display(input_feature_set)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier


model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split


training_inputs, testing_inputs, training_labels, testing_labels = train_test_split(input_feature_set, target_column)
```

```python
testing_labels
```

```python
training_inputs
```

```python
model.fit(training_inputs, training_labels)
```

```python
predictions = model.predict(testing_inputs)
```

```python
from sklearn.metrics import accuracy_score


accuracy_score(testing_labels, predictions)
```

```python
from sklearn.model_selection import learning_curve


(train_sizes_array, train_scores, test_scores) = learning_curve(model, input_feature_set, target_column)
```

```python
train_scores_mean = train_scores.mean(axis = 1)
```

```python
test_scores_mean = test_scores.mean(axis = 1)
```

```python
#plot the learning curve
import matplotlib.pyplot as plt
plt.plot(train_sizes_array, train_scores_mean, label='Training score', marker='o')
plt.plot(train_sizes_array, test_scores_mean, label='Test score', marker='o')
plt.title('Learning Curve')
plt.xlabel('Number of training examples')
plt.ylabel('Score')
plt.legend()
plt.show()
```
