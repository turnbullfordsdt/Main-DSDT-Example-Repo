```python
# import linkage and dendrogram functions from scipy
from scipy.cluster.hierarchy import linkage, dendrogram

# import pandas read_csv function to load in my dataset
from pandas import read_csv

# loading in my dataset
X = read_csv("/workspaces/Main-DSDT-Example-Repo/Machine Learning Fundamentals/no_null_encoded_titanic (1) (1).csv")
# display(X)

# create the fitted hierarchical clustering model
my_model = linkage(X)

# create the dendrogram on the backend
dendrogram(my_model)
```

```python
from scipy.cluster.hierarchy import fcluster
num_clusters_that_I_want = 3
clusters = fcluster(my_model, t=num_clusters_that_I_want, criterion = 'maxclust')
display(clusters)
```

```python
condition = (clusters == 3)

# display(condition)

third_group = X[condition]
```
