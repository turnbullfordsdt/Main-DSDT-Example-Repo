```python
import pandas as pd
df = pd.DataFrame({
    'user_id': [101, 102, 103, 104],
    'tags': [['sports', 'music'], ['cooking', 'books', 'sports'], ['gaming', 'sports'], []]
})


df
```

```python
from numpy import isscalar
```

```python
def any_elements_in_frame_are_not_scalars():
  whether_any_elements_are_not_scalars = False
  for column_name in df:
    column = df[column_name]
    for element in column:
      whether_any_elements_are_not_scalars = not isscalar(element)
      if whether_any_elements_are_not_scalars:
        return whether_any_elements_are_not_scalars
  return whether_any_elements_are_not_scalars
```

```python
def any_elements_in_column_are_not_scalars():
  for element in column:
    whether_any_elements_in_column_are_not_scalars = not isscalar(element)
    if whether_any_elements_in_column_are_not_scalars:
      return whether_any_elements_in_column_are_not_scalars
  return whether_any_elements_in_column_are_not_scalars
```

```python
def other_func(element):
  if isinstance(element, dict):
    return element
  the_element_is_scalar = isscalar(element)
  if the_element_is_scalar:
    element = [element]

  return dict(enumerate(element))
```

```python
def convert_every_element_to_a_dict(column):
  return column.apply(other_func)
```

```python
from pandas import json_normalize
```

```python
def normalize_the_column_into_frame(column):
  semi_flattened_subframe = json_normalize(column)
  return semi_flattened_subframe
```

```python
def prefix_the_names_with_original_name(subframe, name):
  subframe = subframe.add_prefix(name + ".")
  return subframe
```

```python
def horizontally_concat_new_frame_with_old_frame(df, subframe):
  df[subframe.columns] = subframe
  return df
```

```python
def drop_old_name(df, name):
  return df.drop(name, axis=1)
```

```python
# run forever as long as not all elements in frame are scalars
while any_elements_in_frame_are_not_scalars():

  # for every column in our dataframe
  for column_name in df:
    # extract the actual column
    column = df[column_name]

    # convert every element in that column to a dict
    # UNLESS every element is a scalar
    if any_elements_in_column_are_not_scalars():
      updated_column = convert_every_element_to_a_dict(column)
      subframe = normalize_the_column_into_frame(updated_column)
      subframe = prefix_the_names_with_original_name(subframe, column_name)
      df = horizontally_concat_new_frame_with_old_frame(df, subframe)
      df = drop_old_name(df, column_name)
```

```python
df = pd.get_dummies(df)
df
```

```python
cat_tag = '_sports'
same_cat_subframe = df.filter(like=cat_tag)

process_axis = 1
new_aggregate_column = same_cat_subframe.any(axis=process_axis)
aggregate_column_label = 'tags_sports'
df[aggregate_column_label] = new_aggregate_column
df = df.drop(columns = same_cat_subframe)
df
```
