```python
dictified_column = column.copy()

from numpy import isscalar

for index, element in enumerate(column):
    if isscalar(element):
        dictified_element = {0:element}
        dictified_column[index] = dictified_element
    if type(element) == list:
        enumerated_element = enumerate(element)
        dictified_element = dict(enumerated_element)
        dictified_column[index] = dictified_element
```
