```python
def get_year(sample):
    year = sample[:4]
    return year
```

```python
def get_month(sample):
    month = sample[4:6]
    return month
```

```python
def get_day(sample):
    day = sample[6:8]
    return day
```

```python
final_frame['year_last_sold'] = final_frame['date'].apply(get_year).astype(int)

final_frame['month_last_sold'] = final_frame['date'].apply(get_month).astype(int)

final_frame['day_last_sold'] = final_frame['date'].apply(get_day).astype(int)

final_frame = final_frame.drop(columns = 'date')
```
