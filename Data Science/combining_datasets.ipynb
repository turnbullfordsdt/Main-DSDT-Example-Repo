```python
def check_if_string(sample):
    if type(sample) == str:
        return False
    return True

def check_if_not_four_digits(sample):
    sample = str(sample)
    if len(sample) != 4:
        return False
    return True
```

```python
rule1 = (final_frame["yr_built"] < 2015)
rule_filtered_subset = final_frame[rule1]

rule2 = (rule_filtered_subset["yr_built"]).apply(check_if_string)
rule_filtered_subset = rule_filtered_subset[rule2]

rule3 = (rule_filtered_subset["yr_built"]).apply(check_if_not_four_digits)
rule_filtered_subset = rule_filtered_subset[rule3]
```
