```python
from pandas import DataFrame
import pandas as pd

df = pd.DataFrame({
    "word1": ["I", "Python", "write", "is", "makes", "just", "start", "important", "check", "think", "enjoy", "sat", "reading", "went"],
    "word2": ["love", "is", "code", "very", "life", "read", "a", "to", "my", "about", "listening", "on", "a", "for"],
    "word3": ["Python", "fun", "every", "interesting", "easier", "books", "project", "plan", "emails", "ideas", "to", "the", "book", "a"],
    "word4": ["and", "to", "day", "when", "if", "in", "by", "before", "before", "while", "music", "without", "quietly", "walk"],
    "word5": ["coding", "learn", "without", "applied", "used", "the", "planning", "executing", "sleeping", "walking", "during", "moving", "in", "after"],
    "word6": ["", "", "fail", "correctly", "well", "evening", "carefully", "tasks", "tonight", "outside", "work", "", "the", "with"]
})

df
```

```python
encoder_capsule = {}
from sklearn.preprocessing import LabelEncoder

for name in df.columns:
    encoder_capsule[name] = LabelEncoder()
    encoder_capsule[name].fit(df[name])
    df[name] = encoder_capsule[name].transform(df[name])
    display(encoder_capsule[name].classes_)
```

```python
label = df['word6']

input_set = df.drop(columns = ['word6'])
```

```python
from keras import Sequential


model = Sequential()


from keras.layers import Dense, Embedding, Flatten

my_layer = Embedding(input_dim = 14, output_dim = 10)

model.add(my_layer)

my_layer = Flatten()

model.add(my_layer)

my_layer = Dense(units = 14, activation = 'softmax')

model.add(my_layer)
```

```python
model.compile(loss = 'sparse_categorical_crossentropy')
```

```python
model.fit(input_set, label, epochs = 100)
```
